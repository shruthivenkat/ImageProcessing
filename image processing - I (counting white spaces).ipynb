```python
from matplotlib import pyplot as plt
import matplotlib.cm as cm
```

```python
from skimage import data
```

```python
from skimage.feature import blob_dog, blob_log, blob_doh
from math import sqrt
from skimage.color import rgb2grey
import glob
from skimage.io import imread
```

```python
spot_img = glob.glob(r"C:\Users\pc-pc\Documents\Datasets_for_practise\two.png")[0]
im = imread(spot_img, as_grey=True)
plt.imshow(im, cmap=cm.gray)
plt.show()
```

```python
blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold=0.3)
print(blobs_log)

blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
print(blobs_log)

nums = len(blobs_log)
print("Number of stars counted : " ,nums)
```

```python
fig, ax = plt.subplots(1, 1)
plt.imshow(im, cmap=cm.gray)
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+5, color='yellow', linewidth=2, fill=False)
    ax.add_patch(c)
```
